```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module7/data/netflix_titles.csv")
df.drop(columns=['date_added', 'release_year'], inplace=True)
df.fillna('', inplace=True)

df
```

```python
titles = df['title']
data = df.drop(columns=['title'])
```

```python
titles
```

```python
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
df['combined_features'] = df['type'] + ' ' + df['director'] + ' ' + df['cast'] + ' ' + df['country'] + ' ' + df['rating'] + ' ' + df['duration'] + ' ' + df['listed_in'] + ' ' + df['description']
```

```python
df['combined_features']
```

```python
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
tfidf_matrix
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

cosine_sim
```

```python
df_sim = pd.DataFrame(cosine_sim, columns=titles, index=titles)

df_sim
```

```python
title = 'Zombieland'
k = 10

sim_titles = list(df_sim[title].nlargest(k).index)

sim_titles
```

```python
user_titles = list(titles[:3])

user_titles
```

```python
k = 10
final_titles = pd.Series()

for title in user_titles:
  sim_titles = df_sim[title].nlargest(k)

  if len(final_titles) == 0:
    final_titles = sim_titles
  else:
    final_titles = final_titles.add(sim_titles, fill_value=0)

for title in user_titles:
  if title in final_titles:
    final_titles.drop(index=title, inplace=True)


list(final_titles.nlargest(k).index)
```

```python
from scipy.spatial.distance import cosine, jaccard
from scipy.stats import hmean

tfidf_matrix = tfidf_matrix.toarray()

def harmonic_mean_similarity(vec1, vec2):
    cosine_sim = 1 - cosine(vec1, vec2)
    jaccard_sim = 1 - jaccard(vec1 > 0, vec2 > 0)
    return hmean([cosine_sim, jaccard_sim])

similarity_matrix = pd.DataFrame(index=titles, columns=titles, dtype=float)

for i in range(len(titles)):
    for j in range(i, len(titles)):
        if i == j:
            similarity_matrix.iloc[i, j] = 1.0
        else:
            vec1 = tfidf_matrix[i]
            vec2 = tfidf_matrix[j]
            sim_score = harmonic_mean_similarity(vec1, vec2)
            similarity_matrix.iloc[i, j] = sim_score
            similarity_matrix.iloc[j, i] = sim_score

def get_similar_movies(title, k=10):
    sim_scores = similarity_matrix[title].nlargest(k + 1)
    sim_scores = sim_scores.drop(index=title)
    return list(zip(sim_scores.index, sim_scores.values))

similar_movies_with_scores = get_similar_movies('Zombieland', 10)
for movie, score in similar_movies_with_scores:
    print(f"Movie: {movie}, Similarity Score: {score:.4f}")
```
