# Завдання 3
Очистіть дані від викидів. Виведіть очищений датафрейм

```python
columns = df[['hdlngth', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']]

q1 = columns.quantile(0.25)
q3 = columns.quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

mask = (columns >= lower) & (columns <= upper)

cleaned_df = df[mask.all(axis=1)]
```

```python
cleaned_df
```

# Завдання 4
Намалюйте матрицб кореляцій. Для найбільш корельованих ознак намалюйте точкову діаграму.

```python
correlation_matrix = columns.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

most_correlated = correlation_matrix.unstack().sort_values(kind="quicksort", ascending=False)
most_correlated = most_correlated[most_correlated < 1].drop_duplicates()
highest_correlation = most_correlated.idxmax()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=highest_correlation[0], y=highest_correlation[1])
plt.title(f'Scatter Plot: {highest_correlation[0]} vs {highest_correlation[1]}')
plt.xlabel(highest_correlation[0])
plt.ylabel(highest_correlation[1])
plt.show()
```

# Завдання 5
Для 4-ох числових ознак на вибір намалюйте ящики з вусами згрупованими по статі. Подумайте чи є вплив статі на розміри опосума

```python
numeric_columns = ['hdlngth', 'skullw', 'totlngth', 'taill']

plt.figure(figsize=(10, 6))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y='sex', x=column, data=df)
    plt.title(f'Boxplot of {column} by Sex')
    plt.xlabel('Sex')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```
