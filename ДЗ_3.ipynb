# Завдання 6
Перевірте чи правильно розраховано загальну вартість покупки. Виведіть ті рядки де є помилка. Виправте дані в таблиці

```python
df['calculated_sale_amt'] = df['units'] * df['unit_price']
incorrect_sales = df[df['sale_amt'] != df['calculated_sale_amt']]
incorrect_sales
```

```python
df.loc[incorrect_sales.index, 'sale_amt'] = df['calculated_sale_amt']
df.drop(columns=['calculated_sale_amt'], inplace=True)
df
```
